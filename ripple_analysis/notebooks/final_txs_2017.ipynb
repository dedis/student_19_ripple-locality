```python
import pandas as pd
```

```python
data = pd.read_csv('final_txs_2017.txt', header = None, names = ["fee", "hash","sender","success","actualReceiverAmount","5","6","7","8","9","10","flags","actualReceiverCurrency","receiver","date","actualIssuerReceiver"], error_bad_lines = False, warn_bad_lines = False)
```

```python
data.head(10)
```

```python
data.drop(["5","6","7","8","9","10"],axis = 1,inplace = True)
```

```python
data.fee.replace(to_replace='{\"actualReceiverCurrency\": .*', value='', regex=True,inplace = True)
data = data[data.fee != '']
data.fee.replace(to_replace='{"fee": ', value = '', regex = True, inplace = True)
```

```python
data.hash.replace(to_replace=' "hash": "', value = '', regex = True, inplace = True)
data.hash.replace(to_replace='"', value = '', regex = True, inplace = True)
data.hash.replace(to_replace='.*:.*', value='', regex=True,inplace = True)
data = data[data.hash != '']
```

```python
data.sender.replace(to_replace=' "sender": "', value = '', regex = True, inplace = True)
data.sender.replace(to_replace='"', value = '', regex = True, inplace = True)
```

```python
data.success.replace(to_replace=' "success": "', value = '', regex = True, inplace = True)
data.success.replace(to_replace='"', value = '', regex = True, inplace = True)
```

```python
data.actualReceiverAmount.replace(to_replace=' "actualReceiverAmount": "', value = '', regex = True, inplace = True)
data.actualReceiverAmount.replace(to_replace='"', value = '', regex = True, inplace = True)
data.actualReceiverAmount.replace(to_replace='.*:.*', value='', regex=True,inplace = True)
data = data[data.actualReceiverAmount != '']
```

```python
data.flags.replace(to_replace=' "flags": ', value = '', regex = True, inplace = True)
data.flags.replace(to_replace='.*:.*', value='', regex=True,inplace = True)
data.flags.replace(to_replace='""', value='', regex=True,inplace = True)
data = data[data.flags != '']
```

```python
data.actualReceiverCurrency.replace(to_replace=' "actualReceiverCurrency": "', value = '', regex = True, inplace = True)
data.actualReceiverCurrency.replace(to_replace='"', value = '', regex = True, inplace = True)
```

```python
data.receiver.replace(to_replace=' "receiver": "', value = '', regex = True, inplace = True)
data.receiver.replace(to_replace='"', value = '', regex = True, inplace = True)
```

```python
data.date.replace(to_replace=' \"actualIssuerReceiver\": .*', value='', regex=True,inplace = True)
data.date.replace(to_replace=' "date": "', value = '', regex = True, inplace = True)
data.date.replace(to_replace='"', value = '', regex = True, inplace = True)
data.date.replace(to_replace=' date: ', value = '', regex = True, inplace = True)
data.date.replace(to_replace='\d{5,}', value = '', regex = True, inplace = True)
data = data[data.date != '']
```
