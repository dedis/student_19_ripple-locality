# Nodes URLS
We get the nodes URLS from the ripple API

```python
import socket 
import csv
from ripple_api import RippleDataAPIClient
```

```python
api = RippleDataAPIClient('https://data.ripple.com')
topology = api.get_topology()
nb_nodes = topology['node_count']
```

```python
ips = []
i = 0
while i < nb_nodes :
    try:
        ips.append(topology['nodes'][i]['ip']) #get IP address of nodes
        i += 1
    except KeyError:
        i += 1
```

```python
ips
```

```python
urls = []
for x in ips:
    try:
        urls.append(socket.gethostbyaddr(x)[0]) # get URL
    except socket.herror:
        break
```

```python
urls
```

```python
with open('urls.csv', 'w', newline='') as myfile:
     wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
     wr.writerow(urls) #write to disk
```
